### Here i  actual code , that is more readable understandable , optimizeable

```python
import random

while True:
    user_choice = input("Rock, Paper or Scissors? (r/p/s): ").lower()
    comp_choice = random.choice(['r', 'p', 's'])

    print(f"You chose {user_choice}")
    print(f"Computer chose {comp_choice}")

    if user_choice not in ['r', 'p', 's']:
        print("Invalid input")
        continue

    if user_choice == comp_choice:
        print("Draw")

    elif (user_choice == 'r' and comp_choice == 's') or \
         (user_choice == 'p' and comp_choice == 'r') or \
         (user_choice == 's' and comp_choice == 'p'):
        print("You Win")

    else:
        print("Computer Wins")

    play_again = input("Do you want to continue (y/n): ").lower()
    if play_again != 'y':
        break
```
