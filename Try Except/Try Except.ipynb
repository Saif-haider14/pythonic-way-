## 📌 1. Definition

```python
The try-except-else-finally blocks are used to handle errors (exceptions) in Python programs so that the program does not crash.
```

## 📖 2. Description

```python
try → contains code that may cause an error
except → handles the error if it occurs
else → runs only if no error occurs
finally → always runs (whether error happens or not)

👉 This helps make programs safe and user-friendly.
```

## 🧪 3. Basic Syntax

```python
try:
    # risky code
except:
    # handle error
else:
    # runs if no error
finally:
    # always runs
```

## ⚠️ 4. Try Block

```python
The try block contains risky code that might cause an error.

try:
    x = int(input("Enter number: "))
    print(10 / x)

👉 If an error occurs here, Python immediately jumps to except.
```

## 🚫 5. Except Block

```python
The except block is used to handle specific errors.

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")

👉 Python matches the error type and runs the correct block.
```

## ➕ 6. Else Block

```python
The else block runs only when no error occurs.

else:
    print("Everything worked fine")

👉 It is used for code that should run only if try is successful.
```

## 🔚 7. Finally Block

```python
The finally block always executes, no matter what.

finally:
    print("Program finished")

👉 Used for:

Closing files
Releasing resources
Cleanup tasks
```

## 🔄 8. Multiple Except Blocks

```python
You can handle different errors separately.

try:
    x = int(input("Enter number: "))
    print(10 / x)

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")

👉 Python checks from top to bottom and runs the first matching block.
```

## 🎯 9. Full Example (All Blocks Together)

```python

try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ValueError:
    print("❌ Invalid input! Please enter a number.")

except ZeroDivisionError:
    print("❌ Cannot divide by zero.")

else:
    print("✅ Result is:", result)

finally:
    print("🔚 Program execution completed.")
```

```python
🎯 6. Simple Analogy

Think of it like a security guard checking ID cards:

Error = person

except blocks = guards

Each guard checks:
👉 “Is this my type?”

If YES → handles it

If NO → passes to next guard
```
