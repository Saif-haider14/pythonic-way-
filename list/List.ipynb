# 🐍 Python Lists – Complete Beginner Guide

## 📌 1. Definition

```python



A list in Python is a collection of items stored in a single variable.

Lists are ordered

Lists are changeable (mutable)

Lists allow duplicate values
```

```python
fruits = ["apple", "banana", "mango"]
```

## 📖 2. Explanation (Simple Words)

```python


Think of a list like a container (box) where you can store multiple things.

Example:

A student list

A list of marks

A shopping list

You can:

Add items

Remove items

Change items

Access items
```

## 🔹 3. Creating a List

```python

# Empty list
my_list = []

# List with values
numbers = [1, 2, 3, 4]

# Mixed data types
mixed = [1, "hello", 3.5, True]
```

## 🔹 4. Accessing Elements (Indexing)

```python


Index starts from 0

fruits = ["apple", "banana", "mango"]

print(fruits[0])  # apple
print(fruits[1])  # banana
```

```python
print(fruits[-1])  # mango
```

## 🔹 5. Slicing

```python

numbers = [1, 2, 3, 4, 5]

print(numbers[1:4])  # [2, 3, 4]
```

## 🔹 6. Changing Values

```python

fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"
print(fruits)
```

## 🔹 7. Adding Items

```python

# append() → Add at end
fruits.append("grapes")
print(fruits)
```

```python
# insert() → Add at specific position      insert(index , value)
fruits.insert(1, "kiwi")
print(fruits)
```

## 🔹 8. Removing Items

```python

# remove() → Remove by value
fruits.remove("pineapple")
print(fruits)
```

```python
# pop() → Remove by index
fruits.pop(1)

```

```python
🧠 Short Answer

✅ pop() → returns a value

❌ remove() → does NOT return a value

❌ del → does NOT return a value

```

```python
✅ Correct Way to Use del
fruits = ["apple", "banana", "mango"]

del fruits[0]

print(fruits)
```

```python
clear() → Remove all items
fruits.clear()
```

## 🔹 9. List Methods (Important)

```python

Method    	Description
append()	Add item at end
insert()	Add item at index
extend()	Add multiple items
remove()	Remove specific value
pop()	    Remove by index
clear()  	Remove all items
index()	    Find position
count()	    Count occurrences
sort()	    Sort list
reverse()	Reverse list
copy()	    Copy list
```

## 🔹 10. Example of Methods

```python

numbers = [3, 1, 4, 1, 5]

numbers.sort()
print(numbers)  # [1, 1, 3, 4, 5]

numbers.reverse()
print(numbers)  # [5, 4, 3, 1, 1]

print(numbers.count(1))  # 2
```

## 🔹 11. Looping Through List

```python

fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

## 🔹 12. List Comprehension (Important for Industry)

```python


Short way to create lists:

numbers = [1, 2, 3, 4]

squares = [x*x for x in numbers]
print(squares)
```

## 🔹 13. Nested Lists

```python

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix[1][2])  # 6
```

## 🔹 14. Important Concepts

```python

Mutable (Very Important)

Lists can be changed after creation.

a = [1, 2, 3]
a[0] = 100
```

## 💼 15. Interview Questions (with Answers)

```python

Q1: What is a list in Python?

👉 A list is a collection of items that is ordered and mutable.

Q2: Difference between append() and extend()?

👉 append() adds one item
👉 extend() adds multiple items

Q3: What is list slicing?

👉 Extracting a part of a list using index range.

Q4: Are lists mutable?

👉 Yes, lists are mutable.

Q5: What is list comprehension?

👉 A short way to create lists using a single line of code.

Q6: Difference between remove() and pop()?

👉 remove() removes by value
👉 pop() removes by index

Q: Difference between del list and list.clear()?

👉 Answer:

del list deletes the entire list (variable removed)

list.clear() removes all elements but keeps the list
```
