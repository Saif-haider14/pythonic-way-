## Python print() Function: Complete Guide



#### Basic Syntax

## 1. Parameters of print()


### 1.1 *objects



```python
The sep parameter specifies what separates multiple objects.

Default is a single space ' '.

Works only when there are two or more objects.
```

```python
#### Example :
```

```python
name = "Saif"
age = 23

print(name)
print(name, age)
```

### 1.2 sep (Separator)



```python
#### Examples:
```

```python
print("Python", "Course", "2026")        # default sep
print("Python", "Course", "2026", sep="-")
print("Python", "Course", "2026", sep="***")
```

### Common Mistake :

### 1.3 end:

```python
print("hello")
print("saif")
```

```python
print("hello" , end = ' ')
print("saif")
```

```python
print("hello" , end = '')
print("saif")
```

```python
print("hello" , end = ",")
print("saif" , end=",")
print("haider")
```

```python
print("Hello", end=" ")
print("World!")
print("Python", end="---")
print("Course")
```
