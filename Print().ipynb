### File by default ( file=sys.stdout):

```python
with open("saif.txt" , "w") as sf:
    print("hello my friend" , file=sf)

    # with this line of code the output , the output is given to the sf file , and it is printed over there on sf file which means  saif.txt file .
```

## recommended methd 1  (Automatically closes the file dont close the file manually )

```python
# Recommended method
with open("file.txt", "w") as f:
    f.write("Hello")
```

## method 2 (normal open of a file , you have to close it manually )

```python
f = open("file.txt", "w")
f.write("Hello")
f.close()                    # here you manually write  f.close() to close a file 
```

```python
import time

for i in range(5):
    print(i, end=" ")
    time.sleep(1)
```

```python
import time

for i in range(5):
    print(i, end=" ", flush=True)
    time.sleep(1)
```
