```python
In Python programming, a module simply means a file that contains Python code (functions, variables, classes) which you can reuse in other programs.
```

```python
🔹 What is a Module?

A module is just a .py file.

Example:
If you create a file named math_utils.py, that file itself is a module.
```

```python
🔹 Why Modules are Used?
Code reuse (same code bar-bar na likhna)
Better organization
Easy maintenance
Large programs ko small parts me divide karna
```

```python
1. Built-in Modules

Python already provides these modules.

Examples:

math → mathematical operations
random → random values generate karna
datetime → date & time handle karna
```

```python
import math

print(math.sqrt(16))  # output: 4.0
```

```python
❓ Is the dot operator (.) mandatory when using modules?

No, it is not always mandatory.
It depends on how you import the module.
```

```python
🔹 1. Importing the whole module
import math
print(math.sqrt(16))
```

```python
🔹 2. Importing specific functions
from math import sqrt
print(sqrt(16))
```

```python
✔ Here, the dot operator is NOT required
because you imported the function directly.
```

```python
🔹 3. Using an alias
import math as m
print(m.sqrt(16))

✔ Dot operator is still required
(just the module name is shorter)
```

```python
🔹 4. Importing everything (not recommended ❌)
from math import *
print(sqrt(16))
```

```python
❌ Why from module import * is NOT recommended
🔹 1. Name Conflicts (Biggest Problem)

When you import everything, you don’t know which names are coming from where.

from math import *
from random import *

print(sqrt(16))  # Which sqrt? (math or something else?)
```

```python
🔹 Final Rule
import module → use dot (module.function())
from module import function → no dot needed
```

```python
❓ Does import math use more memory than from math import sqrt?

👉 Short answer: NO (not in the way you think).
```

```python
What actually happens
Case 1:
import math
```

```python
Case 2:
from math import sqrt
```

```python
👉 In both cases, Python loads the entire module into memory.

✔ That means:

The full math module is loaded
Not just sqrt
```

```python
🔹 Then what is the difference?

The difference is NOT memory, it’s about namespace (how you access things).

✔ import math

You access using:

math.sqrt(16)
```

```python
Everything stays inside the math namespace
```

```python
✔ from math import sqrt

You access directly:

sqrt(16)
```

```python
-Only sqrt is added to your current namespace
```

```python
🔹 Important Concept (Very Important ⚡)

👉 Python uses module caching:

A module is loaded only once
Even if you import it multiple times

So memory difference is negligible (almost none)
```

```python
🔹 When does memory actually matter?

Memory matters when:

You import very large libraries (like numpy, pandas)
OR you load large datasets/files
```

```python
from numpy import array

👉 still loads most of numpy internally
```

```python
🔹 Your Statement (Corrected)

You said:

importing full module takes more memory than specific function

👉 ❌ Not correct

✔ Correct version:

Both usually load the full module into memory
The difference is only in how names are exposed
```

```python
import math as m 
x = 10.333
print(m.ceil(x))                           # ceil gives you the value greater or equal to the x 
```

```python
import math as m 
x = 10.333
print(m.floor(x))            # floor gives you value less than equal to x 
print(m.floor(-3.4))  # Output: -4
```
