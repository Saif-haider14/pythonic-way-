```python

Think of a dictionary like a real-life dictionary 📖:

Word → Meaning

Key → Value

👉 Example:

"name" → "Ali"

"age" → 20
```

```python
✅ Stores data in key-value pairs

✅ Mutable (can be changed)

❌ No duplicate keys allowed

✅ Keys must be unique

✅ Unordered (but Python 3.7+ keeps insertion order)
```

## 🔄 update() Method Example

```python

📌 What update() Does

👉 It is used to:

Add new key-value pairs

Update existing values
```

## ✅ Example 1: Adding New Items

```python

fruits = {
    "apple": 10,
    "banana": 5
}

fruits.update({
    "mango": 7,
    "orange": 8
})

print(fruits)
```

```python
👉 Output:

{'apple': 10, 'banana': 5, 'mango': 7, 'orange': 8}
```

## ✅ Example 2: Updating Existing Value

```python

fruits = {
    "apple": 10,
    "banana": 5
}

fruits.update({
    "banana": 20
})

print(fruits)
```

```python
👉 Output:

{'apple': 10, 'banana': 20}
```

## ✅ Example 3: Mix (Add + Update Together)

```python

fruits = {
    "apple": 10,
    "banana": 5
}

fruits.update({
    "banana": 15,   # updated
    "grapes": 12    # added
})

print(fruits)
```

```python
👉 Output:

{'apple': 10, 'banana': 15, 'grapes': 12}
```

## 💡 Alternative Syntax

```python


You can also pass another dictionary:

new_data = {"kiwi": 4, "apple": 50}

fruits.update(new_data)
```

# 🐍 Python Dictionary – Interview Questions

```python

Q1: What is a dictionary in Python?

Answer:
A dictionary is a collection of key-value pairs. Keys are unique and immutable, values can be of any data type.

Q2: How is a dictionary different from a list?

Answer:

Feature	    List	                            Dictionary
Access  	By index	                         By key
Order	    Ordered (Python 3.7+)	            Ordered (Python 3.7+)
Structure	Single values	                    Key-value pairs
Use	        Collection                          Mapping
Q3: Can dictionary keys be duplicated?

Answer:
No, dictionary keys must be unique. If you assign the same key again, it updates the value.

Q4: Are dictionaries mutable?

Answer:
Yes, dictionaries are mutable, meaning you can add, remove, or update key-value pairs after creation.

Q5: What is the difference between get() and [] for accessing values?

Answer:

[] → Raises KeyError if key does not exist

get() → Returns None (or a default value) if key does not exist

person = {"name": "Ali"}
print(person.get("age"))  # None
print(person.get("age", 20))  # 20 (default)
Q6: How do you remove an item from a dictionary?

Answer:

pop(key) → Removes the key and returns its value

popitem() → Removes last inserted item

del dict[key] → Deletes a key-value pair

clear() → Deletes all items

Q7: How can you loop through a dictionary?

Answer:

person = {"name": "Ali", "age": 20}

# Loop through keys
for key in person:
    print(key)

# Loop through values
for value in person.values():
    print(value)

# Loop through key-value pairs
for key, value in person.items():
    print(key, value)
Q8: What are nested dictionaries?

Answer:
A dictionary inside another dictionary. Useful to represent complex data.

students = {
    "student1": {"name": "Ali", "age": 20},
    "student2": {"name": "Sara", "age": 22}
}
Q9: How can you copy a dictionary?

Answer:
Use the copy() method:

original = {"name": "Ali"}
new_dict = original.copy()
Q10: What is the purpose of update() method?

Answer:

Add new key-value pairs

Update existing key values

person = {"name": "Ali"}
person.update({"age": 20, "name": "Sara"})
Q11: What is the difference between pop() and popitem()?

Answer:

Method	Removes	Returns
pop(key)	Specific key	Value of the removed key
popitem()	Last inserted key-value	Tuple of key-value
Q12: What types of objects can be keys in a dictionary?

Answer:
Keys must be immutable, like:

String

Number (int, float)

Tuple (if tuple only contains immutable elements)

❌ Lists, sets, or other dictionaries cannot be keys.

Q13: What happens if you use an existing key in update()?

Answer:
The value of that key gets updated. New keys are added.

Q14: Can dictionaries be nested inside lists?

Answer:
Yes! You can have lists of dictionaries and dictionaries inside lists — very common in real-world applications like APIs.

students = [
    {"name": "Ali", "age": 20},
    {"name": "Sara", "age": 22}
]
Q15: Why are dictionaries used in industry?

Answer:

Fast lookups (O(1) complexity)

Store data as key-value pairs

Useful for APIs, configs, JSON data, backend systems, and data processing



```
