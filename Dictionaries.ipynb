```python
# Dictionary Initialization
students = {'Roll no': [101, 102, 103], 'Name': ['Ali', 'Ahmad', 'Amir']}
print(students)
```

```python
students.get('Roll no')
```

```python
# Accessing Items (keys and Values)
list(students.items())
```

```python
# Accessing keys
list(students.keys())
```

```python
# Key based POP
students.pop('Name')
```
