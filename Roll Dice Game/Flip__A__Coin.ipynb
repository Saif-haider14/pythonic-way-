```python
import random

while True:
    choice = input("Flip a Coin? (y/n): ")
    if(choice == 'y'):
        coin = random.choice(['Head' , 'Tail'])   # random.choice is used to pic a  random string  , and you can pass a string , list and tuple to pic up random thing btw those choices
        print(f'{coin}')
        

    elif(choice == 'n'):
        print("Thanks for playing!")
        break

    else:
        print("Invalid Choice!")
```
