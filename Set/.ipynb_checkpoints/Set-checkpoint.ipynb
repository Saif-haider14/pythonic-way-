# 🟢 Python Sets (Beginner Friendly Guide)

## 📌 1. Definition

```python
A set in Python is a collection of unique elements.

👉 This means:

No duplicate values are allowed

The order is not fixed (unordered)
```

## 📖 2. Description

```python
A set is used when:

You want to store unique values only

You don’t care about the order

You want to perform mathematical operations like union, intersection

🔑 Key Features:

Unordered (no indexing)

Mutable (you can add/remove items)

No duplicate values

Written using {}
```

## Example of a set

```python
my_set = {1, 2, 3, 4}
print(my_set)
```

## #❗ Duplicate Values Automatically Removed

```python
my_set = {1, 2, 2, 3, 4, 4}
print(my_set)
```

## ⚠️ Empty Set

```python

empty_set = set()   # Correct
```

```python
empty_set = {}   # This creates a dictionary, NOT a set
```

## 🔧 4. Common Set Methods

### ➤ Add Element

```python
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)
```

### ➤ Remove Element

```python
my_set.remove(2)   # Error if not found
my_set.discard(5)  # No error if not found
```

### ➤ Clear Set

```python
my_set.clear()
```

## 🔗 5. Set Operations

###  ➤ Union (Combine sets)

```python
a = {1, 2}
b = {3, 4}

print(a | b)
```

### ➤ Intersection (Common values)

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a & b)
```

###  ➤ Difference

```python
print(a - b)
```

## 🔍 6. Checking Membership

```python
my_set = {1, 2, 3}

print(2 in my_set)   # True
print(5 in my_set)   # False
```

## ⚡ 7. Important Notes

```python
Sets do NOT support indexing:

my_set[0]   # ❌ Error
```

```python
mixed = {1, "hello", 3.5}
```

## 🎯 8. Real-Life Example

```python

👉 Removing duplicates from a list:

numbers = [1, 2, 2, 3, 4, 4]
unique_numbers = set(numbers)

print(unique_numbers)
```

## 💼 9. Interview Questions

```python

❓ Q1: What is a set in Python?

👉 A set is an unordered collection of unique elements.

❓ Q2: Why are sets unordered?

👉 Because they do not maintain insertion order and are implemented using hashing.

❓ Q3: Difference between remove() and discard()?
Method	Behavior
remove()	Raises error if element not found
discard()	Does not raise error
❓ Q4: Can sets contain duplicate values?

👉 No, duplicates are automatically removed.

❓ Q5: Can we access elements using index?

👉 No, sets are unordered and do not support indexing.

❓ Q6: How to create an empty set?
set()
❓ Q7: Are sets mutable?

👉 Yes, we can add or remove elements.
```
